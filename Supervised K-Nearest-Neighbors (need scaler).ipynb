```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
X,y = mglearn.datasets.make_wave(n_samples=40)
```

```python
Xtrain, Xtest, ytrain, ytest= train_test_split(X,y, random_state=0)
```

```python
reg=KNeighborsRegressor(n_neighbors=3)
```

```python
reg.fit(Xtrain, ytrain).predict(Xtest)
```

```python
reg.score(Xtest, ytest)
```

```python
fig,axes=plt.subplots(1,3, figsize=(15,4), subplot_kw=dict(xticks=[], yticks=[]))
line = np.linspace(-3,3,1000).reshape(-1,1)

for n, ax in zip([1,3,9], axes):
    reg = KNeighborsRegressor(n_neighbors=n)
    reg.fit(Xtrain, ytrain)
    ax.plot(line, reg.predict(line))
    ax.plot(Xtrain, ytrain, '<', c=mglearn.cm2(0))
    ax.plot(Xtest, ytest, '>', c=mglearn.cm2(1))
    ax.set_title(f'{n} neighbor(s)\n train_score:{reg.score(Xtest, ytest)}\n test_score:{reg.score(Xtest, ytest)}')
axes[0].legend(['model prediction', 'training data','test data'])
```
